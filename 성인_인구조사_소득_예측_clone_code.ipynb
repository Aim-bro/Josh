np.log1p는 log(1 + x)를 계산하는 함수로, 0이나 매우 작은 값을 다룰 때 수치적 안정성을 제공하고, 0에 대해서도 안전하게 변환할 수 있다.

np.log는 0이나 음수에 대해 정의되지 않기 때문에, 0이 포함된 데이터를 로그 변환할 때는 np.log 대신 np.log1p를 사용하는 것이 좋다.

```python
t_copy['log_capital_gain'] = np.log1p(t_copy['capital_gain'])
t_copy['log_capital_gain']
```

```python
t_copy[['capital_gain', 'log_capital_gain']].describe()
```

### scaling
- Min-max Scaler
 - 범위가 정해진 값이 필요할 때
 - 아웃라이어에 민감함
- Standard Scaler
 - 평균을 0, 표준편차를 1로 맞추어 정규분포의 특성을 가지도록 함,
 - 아웃라이어에 영향을 덜 받음




```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler
mm_scaler = MinMaxScaler()
st_scaler = StandardScaler()
```

## Encoding, Feature Selection & Extraction


- Encoding
 - 모델이 이해하기 힘든 형태, 애매한 형태, 잘못 학습될 가능성이 있는 Feature에 대해 의미적인 관점에서 변환
 - 종류
   - One-hot Encoding : 범주형 변수에 대해 각 클래스 별 독립적인 Feature 생성(값은 1,0)
   - Label Encoding : 순서가 있는 변수에 대해서 단순 숫자로 넘버링
   - Mean Encoding : 해당 Feature의 각 클래스 별 Target 분포를 바탕으로 Feature 값을 맵핑, overfitting 위험이 높음(Kaggle에서 자주쓰임)
 - 불편한 점이 많아서 직접 Feature를 생성하는게 나을 수 있음
